### Extract the Descent_point information

```python
waypoints = lookup_df['Descent_Points'].values.tolist()
runways = lookup_df['LNDRwy'].values.tolist()
val=[]
for i in range(len(waypoints)):
    for j in range(len(waypoints[i])):
        tmp = list(waypoints[i][j])
        tmp.append(runways[i])
        val.append(tmp)
val
```

```python
wp = pd.DataFrame(val, columns=['wp_name', 'wp_ground', 'wp_elev', 'LNDRwy'])
wp
```
